```python
import numpy as np
import pandas as pd
from mpl_toolkits import mplot3d
import matplotlib.pyplot as plt
```

```python
np.random.seed(121)
n = 100
std = 0.6

balloon = pd.DataFrame({
    'length': np.random.normal(8, std, n),
    'width': np.random.normal(4, std, n),
    'weight': np.zeros(n, dtype='int')}
)

melon = pd.DataFrame({
    'length': np.random.normal(6, std, n),
    'width': np.random.normal(6, std, n),
    'weight': np.zeros(n, dtype='int')}   
)
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')

ax.scatter(balloon['length'], balloon['width'], balloon['weight'], label='balloon', c='r', marker= 7)
ax.scatter(melon['length'], melon['width'], melon['weight'], label= 'melon', c= 'b', marker= 7)

ax.set_xlabel('Length')
ax.set_ylabel('Width')
ax.set_zlabel('Weight')
ax.set_title('3D Graph')

plt.show()
```
